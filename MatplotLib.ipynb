# Multiple lines in 1 plot

```python

x = [1,2,3,4,5,6,7,8]
y = [10,20,30,40,50,60,70,80]
z = [100,200,300,400,500,600,700,800]

plt.plot(x,y,'or--',label = 'X vs Y Value')
plt.plot(x,z,'go-',label = 'X vs Z Value')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Demo Plot - 2')
plt.legend(loc= 0)
```

## Generate  1 to 10 'X' values and generate square values for each numbers and draw a graph between 1 to 10 values vs square of that value, and generate cube value of 1 to 10

```python
x = np.arange (1,11)
y = [i**2 for i in x]
z = [i**3 for i in x]

plt.plot(x,y, "or--",label= 'X vs Y Square Value')
plt.plot(x,z, "go-",label= 'X vs Z Cube Value')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot between individual value vs Square number and cube value')
plt.legend(loc= 0)
```

# Sine Waves

```python
a = np.arange(0,  10* np.pi, 0.1)
b = np.sin(a)

plt.figure(figsize=(10, 4))
plt.grid(True)

plt.plot(a,b)
plt.xlabel('- X axis - ')
plt.ylabel('- Y axis -')
plt.savefig('sin_wave_graph.jpeg')
plt.title('(Sin Wave form)')
```

# Sine Wave Graph with Subplot

```python
a = np.arange(0,  6 * np.pi, 0.1)
b_sin = np.sin(a)
b_cos = np.cos(a)
plt.figure(figsize=(10, 10))


plt.subplot(2,1,1)
plt.plot(a,b_sin,'g-')
plt.title('(sine)')
plt.grid(True)
plt.xlabel('Time (radians)')
plt.ylabel('Amplitude');


plt.subplot(2,1,2)
plt.plot(a,b_cos,'r-')
plt.title('(Cosine)')
plt.grid(True)
plt.xlabel('Time (radians)')
plt.ylabel('Amplitude');
```

# Pie Plot

```python
x = ['Android','Python','.Net','Java','Django','ML','AWS']
y = [20,35,139,124,145,190,100]
# z = [12,22,85,54,75,80,80]
explode = (0,0,0.1,0,0,0,0.2)
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'magenta']


plt.pie(y,labels = x, autopct = '%1.1f%%', explode = explode, 
        
        shadow = True, colors = colors,    pctdistance=0.6,
        labeldistance=1.1);
```
